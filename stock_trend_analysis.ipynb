```python
import pandas as pd
# dataset of Apple stock price in 2025
data = {'Date': ['2025-01-02', '2025-02-03', '2025-03-03', '2025-04-01', '2025-05-01', '2025-06-02', '2025-07-01', '2025-08-01', '2025-09-02', '2025-10-01', '2025-11-03', '2025-12-01'],
    'Close_Price': [182.50, 185.20, 180.80, 189.40, 192.10, 195.50, 191.20, 188.90, 198.30, 202.60, 205.10, 212.40]}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 4))
plt.plot(df['Date'], df['Close_Price'], color='green', marker='o')
plt.title('Apple Stock Price Trend')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
# Calculating the percentage change month-over-month
df['pct_change'] = df['Close_Price'].pct_change() * 100
final_table = df[['Date', 'Close_Price', 'pct_change']]
final_table = final_table.fillna(0)
print("Month on month comparison:")
print(final_table)
```
